## Unsupervised Learning on Country Data

## Abstract
This project explores unsupervised learning techniques to analyze a dataset containing socio-economic and health indicators for 167 countries. By leveraging clustering and dimensionality reduction methods, the project identifies patterns and groups countries based on their similarities in various indicators such as GDP per capita, child mortality, life expectancy, and more.

## Purpose
The objective of this project is to perform unsupervised learning on the country dataset to uncover hidden patterns, identify clusters of countries with similar characteristics, and provide actionable insights into global socio-economic and health disparities.

## Design/methodology/approach
- **Data Preprocessing**: The dataset was cleaned, scaled, and prepared for analysis.
- **Exploratory Data Analysis (EDA)**: Statistical summaries, visualizations, and correlation analysis were performed to understand the data.
- **Dimensionality Reduction**: Principal Component Analysis (PCA) was applied to reduce the dataset's dimensionality while retaining most of the variance.
- **Clustering**: K-Means clustering was used to group countries into clusters based on their socio-economic and health indicators.
- **Evaluation**: The optimal number of clusters was determined using the Elbow Method and Silhouette Scores.

## Findings
- Countries were grouped into three distinct clusters:
    - Cluster 1: High-income countries with low child mortality and high life expectancy.
    - Cluster 2: Low-income countries with high child mortality and low life expectancy.
    - Cluster 3: Moderate-income countries with intermediate socio-economic and health indicators.
- Strong correlations were observed between income, GDP per capita, and health outcomes.
- Outliers in the dataset highlighted unique cases, such as countries with extremely high inflation or GDP per capita.

## Keywords
Unsupervised Learning, Clustering, K-Means, Principal Component Analysis, Socio-Economic Indicators, Health Indicators, Data Analysis, Global Disparities

To explain the optimal cluster size, we can analyze the **Elbow Method** and **Silhouette Scores** graphs:

### Elbow Method:
The Elbow Method graph plots the number of clusters (`k`) against the inertia (sum of squared distances of samples to their closest cluster center). The optimal cluster size is typically at the "elbow point," where the rate of decrease in inertia slows down significantly. From the graph:
- The inertia decreases rapidly for `k = 2` to `k = 3`.
- After `k = 3`, the decrease in inertia becomes more gradual.
- This suggests that `k = 3` is a good choice for the optimal number of clusters.

### Silhouette Scores:
The Silhouette Scores graph measures how well-separated the clusters are. Higher scores indicate better-defined clusters. From the graph:
- The silhouette score is highest at `k = 3` (around 0.30).
- This indicates that the clusters are well-separated and cohesive at this cluster size.

### Conclusion:
Both the Elbow Method and Silhouette Scores suggest that the optimal number of clusters is **3**.
